```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Resmi ykle
para_bgr = cv2.imread("5956322790430853401.jpg")

# Resmi matplotlib ile grntlemek iin BGR'den RGB'ye dntr
para_rgb = cv2.cvtColor(para_bgr, cv2.COLOR_BGR2RGB)

# Resmin boyutlarn al (ykseklik, genilik, kanal says)
h, w, c = para_rgb.shape
print(h, w, c)

# Orijinal resmi grntle
plt.imshow(para_rgb)
plt.axis("off")  # Eksenleri kapat
plt.savefig("madenipara")  # Resmi kaydet

# Resmi gri tonlamaya dntr
para_gri = cv2.cvtColor(para_bgr, cv2.COLOR_BGR2GRAY)

# Gri tonlamal resmi grntle
plt.imshow(para_gri, cmap="gray")

# Gri tonlamal resme Gaussian bulankl uygula
bulanik_para = cv2.GaussianBlur(para_gri, (7, 7), 0)

# Bulank resme ikili eikleme uygula
_, para_bin = cv2.threshold(bulanik_para, 130, 255, cv2.THRESH_BINARY) # Changed 'iskambil_bin' to 'para_bin'

# kili resim zerinde Canny kenar alglama uygula
kenar_para = cv2.Canny(para_bin, 100, 200) # Changed 'kenar_iskambil' to 'kenar_para' and 'iskambil_bin' to 'para_bin'

# Kenarlar grntle
plt.imshow(kenar_para, cmap="gray")
plt.show()

# Kenar resimde konturlar bul
contours, hierarchy = cv2.findContours(kenar_para, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE) # Changed 'kenar_iskambil' to 'kenar_para'

# Konturlar orijinal RGB resme iz
kontorlu_para = para_rgb # Changed 'kontorlu_iskambil' to 'kontorlu_para' and assigned 'para_rgb' instead of undefined 'madeni_parapara'
```
